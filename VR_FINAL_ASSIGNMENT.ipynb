## **IDENTIFYING MISSING DATA VALUES**

```python
#The function utilized below yields any columns containing missing data values within the dataset. 
#True means that the column contains a missing value
#The only column where missing values were detected was in 'geocoded_state'.
df.isnull()
```

##**DELETING MISSING VALUES FROM THE DATASET** 

```python
#The column called 'geocoded_state' contains missing values that will be deleted from the dataset using the function below.
#This transformed the 'geocoded_state' column to contain only true values.
#Additionally, it removed a mysterious state the dataset named "CW", which represented no US state acronym. 
df1 = df.dropna()
df1
```

##**DATASET TRANSFORMATIONS**

```python
#In order to set rules, the quartile ranges for the dependent variable called "inpatient_beds_occupied" were identified.
df['inpatient_beds_occupied'].describe()
```

```python
#Rules will now be created for the dependent variable 'inpatient_beds_occupied'.
def rules(row):
  if row ['inpatient_beds_occupied'] == '12306-541397':
    return 'large'
  elif row ['inpatient_beds_occupied'] == '850-4344.25': 
    return 'small'
  else: 
    return 'medium'
df['inpatient_beds_occupied_MOD'] = df.apply(rules, 1)
df[['inpatient_beds_occupied', 'inpatient_beds_occupied_MOD']].sample(5)
```

```python
#In order to set rules, the quartile ranges for the dependent variable called "inpatient_beds_occupied_by" were identified.
df['inpatient_beds_occupied_by'].describe()
```

```python
#Rules will now be created for the dependent variable 'inpatient_beds_occupied_by'.
def rules(row):
  if row ['inpatient_beds_occupied_by'] =='0-58':
    return 'small' 
  elif row ['inpatient_beds_occupied_by'] =='1421-2731':
    return 'large'
  else: 
    return 'medium'
df['inpatient_beds_occupied_by_MOD'] = df.apply(rules, 1)
df[['inpatient_beds_occupied_by', 'inpatient_beds_occupied_by_MOD']].sample(5)
```

```python
#In order to set rules, the quartile ranges for the dependent variable called "staffed_adult_icu_beds" were identified.
df['staffed_adult_icu_beds'].describe()
```

```python
#Rules will now be created for the dependent variable 'staffed_adult_icu_beds'.
def rules(row):
  if row ['staffed_adult_icu_beds'] == '78-504':
    return 'small'
  elif row ['staffed_adult_icu_beds'] == '505-1007': 
    return 'medium'
  else: 
    return 'large'
df['staffed_adult_icu_beds_MOD'] = df.apply(rules, 1)
df[['staffed_adult_icu_beds', 'staffed_adult_icu_beds_MOD']].sample(5)   
```
